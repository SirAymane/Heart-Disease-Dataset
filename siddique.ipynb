### Heart Disease

```python
# Imports

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read csv dataset
dataset=pd.read_csv('./data/heart_clean.csv')
dataset

```

### Relación de hombres y mujeres según escala de dolor.

En este primero apartado, se pretende estudiar la relación entre el recuento de hombres i mujeres según la escala de dolor. 

Se pretende estudiar la escala de dolor de pecho con más afectados y el sexo que más afectado está. 

Hay cuatro tipos de dolor de pecho:

- 0 -> Angina típica
- 1 -> Angina atípica
- 2 -> Dolor torácico no anginoso
- 3 -> Asintomático

Los columnas con las que vamos a trabajar en este apartado son el chest pain type ("cp"), el sexo ("sex") i el recuento de hombres i mujeres según escala de dolor.

Tal y como se puede observar en la gráfica, la anguina típica és la que tiene más recuento de personas, tanto de hombres como mujeres. Aparte, cabe destacar que hay más hombres en los recuentos que mujeres. Esto se debe a que en el estudio, el número de hombres es mayor al número de mujeres.

```python
# Countplot

# Set the plot size
plt.figure(figsize=(10,8)) 

f = sns.countplot(data=dataset, x="cp", hue="sex")

# Title of the plot
plt.title("Escala de dolor de pecho según sexo")

# Set tick Labels for x-axis
f.set_xticklabels(['0-Angina típica', '1-Angina atípica', '2-Dolor torácico no anginoso', '3-Asintomático'])

# Labels for "x" and "y"
plt.ylabel('Recuento')
plt.xlabel('Tipo de dolor de pecho')

# Show plot
plt.show()
```

A continuación, podemos observar el recuento de hombres y mujeres.

```python
# Counplot
sns.countplot(x='sex',data=dataset,palette="dark")

# Labels
plt.xlabel('Sexo')
plt.ylabel('Recuento')

# Show plot
plt.show()
```

### Relacion sexo y frecuencia de ataques en el corazón

De igual manera, podemos hacer el estudio de la frequencia del ataque de corazón según el sexo. Anteriormente hemos observado el reccuento de hombres y mujeres según la escala de dolor pecho. 

En este caso, observaremos si están afectados o no, y haremos el estudio con la columna "target", que con el parámetro 0 indica que no afectado y con el parámetro 1 indica que está enfermo.

```python
# Crosstab
pd.crosstab(dataset.target, dataset.sex).plot(kind="bar",
                                    figsize=(8, 5),
                                    color=["lightgray", "lightblue"])

# Title of the plot
plt.title("Frecuencia de dolor de corazón según sexo")

# Labels
plt.xlabel("Heart disease")
plt.ylabel("Cantidad")

# Legend
plt.legend(["Hombres", "Mujeres"])

# Tick locations for the x-axis
plt.xticks(rotation=0)
```

### Relacion de la edad y la frecuencia de ataques en el corazón


En este otro apartado, se pretende estudiar la relación de la edad y la frecuencia de ataques en el corazón.

Según los estudios de una investigación del Colegio Americano de Cardiología, los ataques cardíacos se dan cada vez más en adultos muy jóvenes, menores de 40 años.

Por ello, vamos a observar en forma de gráfica en qué edad és donde hay más enfermos del ataque de corazón.

Tal y como se puede observar en la gráfica de abajo, hay más afectados en el rango de edad 40 - 54 años. 

Una de las razones es debido a que los adultos mayores están más expuestos a sus factores, como la obesidad, hipertensión y tabaquismo.

Por otro lado, en el rango de 56-65 años es donde hay más personas sanas.

```python
# Corosstab
pd.crosstab(dataset.age,dataset.target).plot(kind='bar',figsize = (20, 8))

# Title of the plot
plt.title('Frecuencia de ataques en el corazón según la edad')

# Labels
plt.xlabel('Edad')
plt.ylabel('Frecuencia de ataques en el corazón')

# Show plot
plt.show()
```

Podemos observar el rango de hombres y mujeres según edad

```python
# Boxenplot

# Set plot size
plt.figure(figsize=(6,4)) 

sns.boxenplot(data=dataset, x="age", y="sex")

# Labels
plt.xlabel('Edad')
plt.ylabel('Sexo')

```

### Relación nivel de colesterol y edad, según sexo

En este último apartado, se pretende estudiar el nivel de colesterol según la edad. A medida que las mujeres y hombres envejecen, los niveles de colesterol suben. 

Tal y como podemos observar en el gráfico, los niveles de colesterol de colesterol de las personas son bastante altos. Se uede observar que las mujeres tienen niveles de colesterol más altos que los hombres. La edad en el que el nivel de colesterol es más alto es a partir de los 50 años.

El nivel de colesterol óptimo para las personas menores de 19 años es menor de 170 mg/dl y en las personas mayores de 20 años, el nivel de colesterol total debe oscilar entre los 125-200 mg/dl.

```python
# Scatterplot

sns.scatterplot(data=dataset, x="age", y="chol", hue="sex", palette="ch:r=-.2,d=.3_r")

# Labels
plt.xlabel('Edad')
plt.ylabel('Colesterol sérico en mg/dl')
```

### Conclusion

En conclusión, cabe destacar que hay algunos factores de riesgo de la enfermedad cardíaca que no se pueden cambiar, como la edad (los hombres de 45 años o más y las mujeres de 55 años o más tienen un mayor riesg), el sexo (algunos factores de riesgo pueden afectar de manera diferente a las mujeres comparados con los hombres), o la historia familiar (hay un mayor riesgo si hay un familiar cercano que tuvo una enfermedad cardíaca a una edad temprana).

No obstante, hay muchas medidas para reducir las posibilidades de tener enfermedades del corazón. Algunas de estas, son las siguientes:
- Mantener sus niveles de colesterol y triglicéridos bajo control
- Mantener un peso saludable
- Hacer ejercicio regularmente
- No fumar
- Controlar el estrés
- Controlar la diabetes
