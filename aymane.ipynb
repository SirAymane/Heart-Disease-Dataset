# Segundo mito sobre la salud del corazón

## “La insuficiencia cardíaca significa que el corazón deja de latir”

Conexió entre ST depression i el máxim heart rate - Una ST depression es una anomalia que muestra un electrocardiograma (ECG), se le llama así ya que se detecta un aumento y una bajada (depressió) de forma anormal en los registros de un ECG.

En esta gráfica se analizan i se comparan los resultados que daría una ECG.

```python
# Reading columns used in this graph.
df2_ay = df_ay[["thalach", "oldpeak", "target"]]

# Creating data structure based on the columns used.
g2 = sns.relplot(
    data=df2_ay, palette="dark",
    x="thalach", y="oldpeak", hue="target",
    col_wrap=None, height=4, aspect=2, kind="scatter"
)

# Setting up label and range
g2.set(ylim=(0.5, 5))
g2.map(plt.axhline, y=0, zorder=1)
g2.set_axis_labels("Maximum Hearth Rate", "ST depression")
```

Analizando los resultados, encoontramos que los enfermos se encuentran a partir de 120 a 180, además de una bajada entre 0.5 y 2.5.

El corazón deja de latir de repente durante un paro cardíaco, no durante una insuficiencia cardíaca. Cuando sufre una insuficiencia cardíaca, el corazón sigue funcionando, pero no bombea sangre tan bien como debería. 

# Tercer Mito sobre la salud del corazón

## “Las cardiopatías son frecuentes en mi familia, así que no puedo hacer nada al respecto”.

Tomamos como dato varialbe el colesterol. Todo y que haya ciertas personas tienen una predisposición intrínsica a una tendencia de los niveles de colesterol. 

Es un factor importante, ya que el colesterol es un lípido que necesita unirse a proteínas transportadoras conocidas como lipoproteínas para circular por la sangre, entonces se encuentra presente constantemente.

De igual forma para llegar a conocer como afecta, veremos los resultados mínimos y máximos de colesterol, comparados con el nivel de dolor de las personas afectadas.

```python
# Reading columns used in this graph.
df3_ay = df_ay[[ "chol", "cp", "target"]]

# Displaying data structure selecting the ranges we are interested on.
g3 = sns.jointplot(data=df3_ay, x="chol", y="cp", kind="hist", color="red", xlim=(125, 325))
```

Realmente hay casos de enfermedades cardiovasculares que son hereditarias, pero esto no implica que se sufrirá la enfermedad siempre, sino más bien que hay un aumento de riesgo de sufrirla.

En este caso reducir los niveles de colesterol implicará que en los pocos casos que pueda ocurrir, el dolor será menor.

A pesar de tener cieta susceptibilidad genética, los estudios demuestran que en la gran mayoría de casos es posible llevar a cabo una buena dieta, entre otros factores de estilo de vida, para reducir el colesterol y practicamente evitar totalmente la enfermedad a pesar de la carga genética que se pueda poseer.
